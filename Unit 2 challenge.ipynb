```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_excel('35478-0001-Data.xlsx') #I used the General Social Survey, 2012 Merged Data
```

```python
#this is visualization of plot showing the avg income of response by age
df_mean = df.groupby('AGE').mean() #claculate data mean by age
plt.plot(df_mean['RINCOME'], color = 'purple')
plt.ylabel('Avg. Income')
plt.xlabel('Age')
plt.title('Avg. Income by Age')
plt.show()
```

```python
#this is visualization of scatter showing the income and age relationship by sex
male = df[df['SEX']==1] #filtering data by sex, male only
female = df[df['SEX']==2] #filtering data by sex, female only
plt.scatter(
    x = male['AGE'], y = male['RINCOME'],
    color = 'blue', marker = 'x', s = 10
)
plt.scatter(
    x = female['AGE'], y = female['RINCOME'],
    color = 'pink', marker = 'o', s = 10
)
plt.ylim([0, 15]) #as most of the response has Income up to 15 filtering the outlayer
plt.ylabel('Income')
plt.xlabel('Age')
plt.title('Income and Age relationship by Sex')
plt.show()
```

```python
#as the values are very similar and it is hard to figure out all on one plot, divided to 2 deifeent plots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(
    x = male['AGE'], y = male['RINCOME'],
    color = 'blue', marker = 'x', s = 10
)
plt.ylim([0, 15]) #as most of the response has Income up to 15 filtering the outlayer
plt.ylabel('Income')
plt.xlabel('Age')
plt.title('Income and Age relationship - Male')

plt.subplot(1, 2, 2)
plt.scatter(
    x = female['AGE'], y = female['RINCOME'],
    color = 'pink', marker = 'o', s = 10
)
plt.ylim([0, 15]) #as most of the response has Income up to 15 filtering the outlayer
plt.ylabel('Income')
plt.xlabel('Age')
plt.title('Income and Age relationship - Female')
plt.show()
```

```python
#this is visualization of response age distribution
plt.hist(df['AGE'],bins=18, color='purple')
plt.title('Age Histogram')
plt.xlabel('Age')
plt.ylabel('# of response')
plt.grid(True)
plt.show()
```
